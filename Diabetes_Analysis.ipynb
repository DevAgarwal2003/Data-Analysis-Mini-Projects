```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('diabetes.csv')
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.describe().round(2)
```

```python
pos = []
for x in df['Outcome']:
    if(x == 'tested_positive'):
        pos.append(x)
percent_pos = (len(pos)/len(df['Outcome']))*100
np.round(percent_pos,2)
```

```python
sns.boxplot(x = 'Outcome', y = 'DiabetesPedigreeFunction', data = df)
```

```python
boolean = []
for x in df['Outcome']:
    if(x == 'tested_positive'):
        boolean.append(0)
    else:
        boolean.append(1)
df['outcome_bool'] = boolean
df
```

```python
df.corr()
```

```python
df.corr()['outcome_bool']
```

```python
list = [] 
for x in df['BloodPressure']:
    if(x<80):
        list.append('Normal diastolic')
    if(80<=x<=89):
        list.append('At Risk')
    if(x>=90):
        list.append('High BP')
df['Status'] = list
df
```
